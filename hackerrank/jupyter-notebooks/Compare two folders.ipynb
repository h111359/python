```python
import os

def compare_folders(folder1, folder2):
    for root, dirs, files in os.walk(folder1):
        relative_path = os.path.relpath(root, folder1)
        compare_dirs = os.path.join(folder2, relative_path)
        if not os.path.isdir(compare_dirs):
            print(f"Directory '{compare_dirs}' not found in the second folder.")
            continue

        # Compare directory names
        for dir_name in dirs:
            compare_dir = os.path.join(compare_dirs, dir_name)
            if not os.path.isdir(compare_dir):
                print(f"Directory '{compare_dir}' not found in the second folder.")
            else:
                print(f"Both folders have directory: '{compare_dir}'")

        # Compare file names and sizes
        for file_name in files:
            file_path1 = os.path.join(root, file_name)
            file_path2 = os.path.join(compare_dirs, file_name)
            if not os.path.isfile(file_path2):
                print(f"File '{file_path2}' not found in the second folder.")
            else:
                size1 = os.path.getsize(file_path1)
                size2 = os.path.getsize(file_path2)
                if size1 != size2:
                    print(f"File '{file_path2}' has different size in the second folder.")
                else:
                    print(f"Both folders have file: '{file_path2}' with the same size.")


```
